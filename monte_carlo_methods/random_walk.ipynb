### **Random Walk (1 Dimension)**  
> Realize it by considering N=10, 100, 10000, 100000

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def random_walk_1D(walk_num: int) -> int:
    """
    Implement 1D random walk, 
    if x less than 0.5, walk += 1 
    else, walk -= 1
    """
    walk = 0
    for _ in range(walk_num):
        x = np.random.uniform(0, 1, 1)
        if x < 0.5:
            walk += 1
        else:
            walk -= 1
    return walk

def plot_random_walk_1D(N_list: int):
    """
    Subplots 1D Random Walk on different Walks
    """
    for idx, N in enumerate(N_list):
        walk_list = []
        for walk_num in range(N):
            walk = random_walk_1D(walk_num)
            walk_list.append(walk)
        plt.subplot(2, 2, idx+1)
        plt.plot(range(N), walk_list)

    plt.suptitle(f"1D Random Walk consdering walk = {N_list}", fontweight='bold')

N_list = [10, 100, 1000, 10000]
plot_random_walk_1D(N_list)
```

### **Random Walk (2 Dimension)**

**Algorithm:**  
1. **Input**:  
   - `n`: Number of steps in each walk  
   - `N`: Number of random walks  

2. **For each walk (N times):**  
   3. **Initialize**: Set the starting position to the origin \((x = 0, y = 0)\).  
   4. **Loop through `n` steps**:  
      - **Randomly choose** a direction from `[N, S, E, W]`.  
      - **Update coordinates based on direction**:  
        - If step is **N**: Increment `y` by `1`.  
        - If step is **S**: Decrement `y` by `1`.  
        - If step is **E**: Increment `x` by `1`.  
        - If step is **W**: Decrement `x` by `1`.  
   5. **Display Results**:  
      - **Final coordinates** \((x, y)\) after `n` steps.  
6. Plot the result




```python
import random 

def random_walk_2D(num_blocks: int) -> tuple:
    x, y = (0, 0)
    for _ in range(num_blocks):
        dx, dy = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy
    
    return x, y

def iterate_random_walk_2D(num_walks: int, num_blocks: int) -> tuple:
    x_walk = []
    y_walk = []
    for _ in range(num_walks):
        x, y = random_walk_2D(num_blocks)
        
        x_walk.append(x)
        y_walk.append(y)
    
    return x_walk, y_walk

def line_plot(x: list, y: list):
    plt.plot(x, y)
    plt.annotate("(0, 0)", (0, 0), color='blue')
    plt.title("Random Walk in 2 Dimensional Plane", fontweight='bold')
    plt.xlabel("X co-ordinate")
    plt.ylabel("Y co-ordinate")

NUM_WALKS = 100
NUM_BLOCKS = 12 

x_walk: list
y_walk: list
x_walk, y_walk = iterate_random_walk_2D(NUM_WALKS, NUM_BLOCKS)
line_plot(x_walk, y_walk)
```
